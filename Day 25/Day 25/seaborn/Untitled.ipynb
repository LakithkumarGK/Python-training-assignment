```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Example dataset
data = {
    'Size': [50, 60, 70, 80, 90],
    'Price': [200, 300, 400, 500, 600]
}

# Convert data to a pandas DataFrame
df = pd.DataFrame(data)

# Create a scatter plot with regression line
sns.lmplot(x="Size", y="Price", data=df)

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
tips = sns.load_dataset("tips")

# Create an lmplot
sns.lmplot(
    x="total_bill",  # X-axis variable
    y="tip",         # Y-axis variable
    hue="sex",       # Group by the "sex" column (color-coded)
    data=tips,       # DataFrame with the data
    markers=["o", "x"], # Custom markers for groups
    aspect=1.5       # Aspect ratio of the plot
)

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a violin plot
sns.violinplot(x="day", y="total_bill", hue="sex", data=tips, split=True)

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate a random dataset
data = np.random.rand(5, 5)  # 5x5 matrix of random numbers

# Create a heatmap
sns.heatmap(data, annot=True, cmap="coolwarm", linewidths=0.5)

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
tips = sns.load_dataset("tips")

# Create a boxplot
sns.boxplot(x="day", y="total_bill", data=tips)

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
tips = sns.load_dataset("tips")

# Create a displot
sns.displot(tips["total_bill"], kde=True)

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
tips = sns.load_dataset("tips")

# Create a jointplot
sns.jointplot(x="total_bill", y="tip", data=tips, kind="scatter")

# Show the plot
plt.show()
```
